## 1. and 2. Loading the data and create subset for december 2006 to november 2009

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('house_Power_Consumption_preprocessed.csv')
```

```python
# Histogram of Global Active Power
plt.figure(figsize=(10, 6))
sns.histplot(df['Global_active_power'], kde=True)
plt.title('Histogram of Global Active Power')
plt.xlabel('Global Active Power (scaled)')
plt.ylabel('Frequency')
plt.show()
```

## 4. creating a time series plot

```python
df['Datetime'] = pd.to_datetime(df['Datetime'])

# Set 'Datetime' as the index
df.set_index('Datetime', inplace=True)

# Aggregate data to daily frequency
df_daily = df.resample('D').mean()

# Plotting the daily time series
plt.figure(figsize=(15, 10))

# Plot for Global Active Power
plt.subplot(3, 1, 1)
plt.plot(df_daily.index, df_daily['Global_active_power'], label='Daily Global Active Power', color='blue')
plt.title('Daily Time Series of Global Active Power')
plt.xlabel('Date')
plt.ylabel('Global Active Power (scaled)')
plt.legend()
plt.grid(True)
```

## 5. Creating plot for submetering

```python
# Plotting the daily time series for sub-metering
plt.figure(figsize=(15, 6))

# Plot for Sub Metering 1
plt.plot(df_daily.index, df_daily['Sub_metering_1'], label='Sub_metering_1', color='blue')

# Plot for Sub Metering 2
plt.plot(df_daily.index, df_daily['Sub_metering_2'], label='Sub_metering_2', color='green')

# Plot for Sub Metering 3
plt.plot(df_daily.index, df_daily['Sub_metering_3'], label='Sub_metering_3', color='red')

plt.title('Daily Time Series of Sub Metering')
plt.xlabel('Date')
plt.ylabel('Energy Sub Metering (scaled)')
plt.legend()
plt.grid(True)
plt.show()
```

## 6. Creating multiple plots

#### Scatter plot

```python
# Aggregate data to daily frequency
df_daily = df.resample('D').mean()  # Use mean for daily aggregation

# Plotting the daily scatter plot for Global Active Power vs. Voltage
plt.figure(figsize=(10, 6))
plt.scatter(df_daily['Global_active_power'], df_daily['Voltage'], alpha=0.5, color='blue')
plt.title('Scatter Plot of Daily Global Active Power vs. Voltage')
plt.xlabel('Global Active Power (Daily Mean)')
plt.ylabel('Voltage (Daily Mean)')
plt.grid(True)
plt.show()
```

#### Histogram

```python
import seaborn as sns

# Histogram for Global Active Power
plt.figure(figsize=(10, 6))
plt.hist(df['Global_active_power'].dropna(), bins=30, color='blue', edgecolor='black')
plt.title('Histogram of Global Active Power')
plt.xlabel('Global Active Power')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

#### Pie chart

```python
# Pie Chart for Sub Metering (example: one day data)
daily_sum = df[['Sub_metering_1', 'Sub_metering_2', 'Sub_metering_3']].sum()
plt.figure(figsize=(8, 8))
plt.pie(daily_sum, labels=daily_sum.index, autopct='%1.1f%%', colors=['blue', 'green', 'red'])
plt.title('Pie Chart of Sub Metering')
plt.show()
```

#### Box plot

```python
# Box Plot for Global Active Power
plt.figure(figsize=(10, 6))
sns.boxplot(y=df['Global_active_power'], color='blue')
plt.title('Box Plot of Global Active Power')
plt.ylabel('Global Active Power')
plt.grid(True)
plt.show()
```

#### Bar Graph

```python
# Bar Graph for Mean Sub Metering
daily_mean_sub_metering = df[['Sub_metering_1', 'Sub_metering_2', 'Sub_metering_3']].mean()
plt.figure(figsize=(10, 6))
daily_mean_sub_metering.plot(kind='bar', color=['blue', 'green', 'red'])
plt.title('Mean Sub Metering')
plt.xlabel('Sub Metering Type')
plt.ylabel('Mean Value')
plt.grid(True)
plt.show()
```

#### Count plot

```python
# Count Plot for Day of Week
plt.figure(figsize=(10, 6))
sns.countplot(x='Day_of_Week', data=df, palette='viridis')
plt.title('Count Plot of Day of the Week')
plt.xlabel('Day of Week')
plt.ylabel('Count')
plt.grid(True)
plt.show()
```

#### Heatmap

```python
# Heatmap of correlations between numeric features
plt.figure(figsize=(12, 10))
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Heatmap of Feature Correlations')
plt.show()
```

#### Distribution Plot

```python
# Distribution Plot for Global Active Power
plt.figure(figsize=(10, 6))
sns.histplot(df['Global_active_power'].dropna(), kde=True, color='blue')
plt.title('Distribution Plot of Global Active Power')
plt.xlabel('Global Active Power')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```

#### Joint Plot

```python
# Joint Plot of Global Active Power and Voltage
plt.figure(figsize=(10, 6))
sns.jointplot(x='Global_active_power', y='Voltage', data=df, kind='scatter', color='blue')
plt.title('Joint Plot of Global Active Power and Voltage')
plt.show()
```
